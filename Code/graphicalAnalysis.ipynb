# Import Visualization Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

# Helper Function

```python
def diff(list1, list2):
  difference = []
  zip_object = zip(list1, list2)
  for list1_i, list2_i in zip_object:
    difference.append(list1_i-list2_i)
  return difference
```

# **Analysis for Figure 1**

```python
!cp "/content/drive/MyDrive/LinkedIn Project/Dataset/ageRangePlot.csv" .
!cp "/content/drive/MyDrive/LinkedIn Project/Dataset/seniorityPlot.csv" .
```

```python
ageData = pd.read_csv("ageRangePlot.csv")
seniorityData = pd.read_csv("seniorityPlot.csv")
```

```python
plt.rcParams['figure.figsize'] = [9, 8]
```

```python
ages = ['18 to 24', '25 to 34', '35 to 54', '55+']
```

```python
female_age_dist = ageData[ageData["Gender"] == "Female"]["Ratio"].to_list()
male_age_dist = ageData[ageData["Gender"] == "Male"]["Ratio"].to_list()
```

```python
female_low = ageData[ageData["Gender"] == "Female"]["Low"].to_list()
female_high = ageData[ageData["Gender"] == "Female"]["High"].to_list()
male_low = ageData[ageData["Gender"] == "Male"]["Low"].to_list()
male_high = ageData[ageData["Gender"] == "Male"]["High"].to_list()
```

```python
female_low = diff(female_age_dist, female_low)
female_high = diff(female_high, female_age_dist)
male_low = diff(male_age_dist, male_low)
male_high = diff(male_high, male_age_dist)
```

```python
xi = list(range(len(ages)))

xi_female = diff(xi, [0.04, 0.04, 0.04, 0.04])
xi_male = diff(xi, [-0.04, -0.04, -0.04, -0.04])
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(xi_female, female_age_dist, 'o-', c="deeppink")
plt.errorbar(xi_female, female_age_dist, yerr=(female_low, female_high), fmt="o", color="deeppink")
plt.plot(xi_male, male_age_dist, 'o-', c="darkgoldenrod")
plt.errorbar(xi_male, male_age_dist, yerr=(male_low, male_high), fmt="o", color="darkgoldenrod")
plt.xticks(xi, ages, fontsize=14, c="black")
plt.yticks(fontsize=13, c="black")
plt.title('', fontsize=20)
plt.xlabel('Age Range', fontsize=16, c="black")
plt.ylabel('Proportion', fontsize=16, c="black")
one_patch = mpatches.Patch(color='deeppink', label='Female')
two_patch = mpatches.Patch(color='darkgoldenrod', label='Male')
plt.legend(handles=[one_patch, two_patch], prop={'size': 12})
plt.savefig("Fine_Age_Distribution.svg")
plt.show()
```
